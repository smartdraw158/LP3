```python
import matplotlib.pyplot as plt 
def function(x):
  return (x + 3)**2 
def gradient(x):
  return 2 * (x + 3)
```

```python
# Gradient Descent Algorithm
```

```python
def gradient_descent(starting_x, learning_rate, num_iterations):
    x = starting_x
    x_values = [x]  # Track x values for visualization
    for i in range(num_iterations):
        grad = gradient(x)
        x = x - learning_rate * grad
        x_values.append(x)
        if abs(grad) < 1e-6:  # Stop if gradient is very small
            break
    return x, x_values
```

```python
# Parameters for Gradient Descent
starting_x = 2         # Starting point
learning_rate = 0.1    # Learning rate (alpha)
num_iterations = 100   # Maximum number of iterations
```

```python
# Perform Gradient Descent
min_x, x_values = gradient_descent(starting_x, learning_rate, num_iterations)
```

```python
# Print the result
print("The local minimum occurs at x =", min_x)
```

```python
# Plotting the function and the steps of gradient descent
x_range = [i for i in range(-10, 5)]
y_range = [function(x) for x in x_range]
```

```python
plt.plot(x_range, y_range, label="y = (x + 3)^2")
plt.scatter(x_values, [function(x) for x in x_values], color='red', label="Gradient Descent Steps")
plt.title("Gradient Descent to find the local minimum of y = (x + 3)^2")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
